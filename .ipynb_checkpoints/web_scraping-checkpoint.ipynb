```python
# Import libraries
from bs4 import BeautifulSoup
import requests
```

```python
# Collect first page of title's list
link = 'https://news.detik.com/'
request = requests.get(link)
content = request.content
```

```python
soup = BeautifulSoup(content,"html.parser") 
print(soup)
```

```python
isi = soup.find_all("a", {"class" : "media__link"})
isi_text = ''

for title in isi:
    isi_text += title.get_text(strip=True) + '\n'
    
print(isi_text)
```

```python
import datetime
  
filename = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
f = open(f"./web scraping/{filename}.txt", "x")
f.write(isi_text)
f.close()

#open and read the file after the appending:
f = open(f"./web scraping/{filename}.txt", "r")
print(f.read())
```
